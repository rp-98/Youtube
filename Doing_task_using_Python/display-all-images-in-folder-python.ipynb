code to: Display all the images from a folder

steps
1. Access the folder/ path ----os
2. Read the images --- opencv---cv2--imread()
3. display ---matplotlit--- imshow()

```python
import os
import cv2
import matplotlib.pyplot as plt
```

```python
#path
path='pets'
images=os.listdir(path)
type(images)
len(images)
```

```python
img_data=[]
for img in images:
    img_arr=cv2.imread(os.path.join(path,img))
    img_data.append(img_arr)
    
```

```python
plt.figure(figsize=(10,10))
for i in range(len(img_data)):
    plt.subplot(6,4,i+1)
    plt.imshow(img_data[i])
```
